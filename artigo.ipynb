```python
import pandas as pd
import pingouin as pg
data = pd.read_excel('data/idade.xlsx')
data
```

Retiranndo as columns que não tem resultados e retirando os NA

```python
data.columns
a_retirar = [
    '@12.Emrelaçãoassuascaracterísticasporfavorresponda',
    '@13.Pararesponderasquestõesabaixoconsidereseuprojetoatua',
    '@14.Emrelaçãoaoseudiaadiadotrabalho',
    '@15.Emrelaçãoasuasatisfaçãocomotrabalho'

]

data.drop(columns= a_retirar, inplace=True)

data = data.dropna(axis=0)
data.columns
```

Agora eu irei aplicar o teste MWU

The Mann-Whitney U test is a nonparametric test of the null hypothesis that the distribution underlying sample x is the same as the distribution underlying sample y. It is often used as a test of difference in location between distributions.

Esse teste foi escolhido porque as distribuições nao sao paramétricas.
Além disso, as respostas são independentes.

O teste diz que se p-val<0.05, as distribuições são semelhantes, do contrário, são diferente.

```python
pg.mwu(data_masculino['Cinismo'], data_feminino['Cinismo'])
```

Cinismo é marginalmente  igual 

```python
pg.mwu(data_masculino['Exaustão'], data_feminino['Exaustão'])
```

exaustao é  diferente

```python
pg.mwu(data_masculino['Eficacia'], data_feminino['Eficacia'])
```

eficacia é bem diferente

```python
pg.mwu(data_masculino['Inst.Equipe'], data_feminino['Inst.Equipe'])
```

Inst.Equipe é marginalmente igual

```python
pg.mwu(data_masculino['Inst.Tarefa'], data_feminino['Inst.Tarefa'])
```

Inst.tarefa é igual

```python
pg.mwu(data_masculino['Inst.Tecnica'], data_feminino['Inst.Tecnica'])
```
