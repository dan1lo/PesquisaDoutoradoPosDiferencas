Analise dos data entre o burnout e instabilidade

```python
import pandas as pd
data = pd.read_excel('data/idade.xlsx')
data
```

Retiranndo as columns que não tem resultados e retirando os NA

```python
data.columns
a_retirar = [
    '@12.Emrelaçãoassuascaracterísticasporfavorresponda',
    '@13.Pararesponderasquestõesabaixoconsidereseuprojetoatua',
    '@14.Emrelaçãoaoseudiaadiadotrabalho',
    '@15.Emrelaçãoasuasatisfaçãocomotrabalho'

]

data.drop(columns= a_retirar, inplace=True)

data = data.dropna(axis=0)
data.columns
```

```python
def mudarSexo(sexo):
    if sexo == 'Masculino':
        return 1
    elif sexo == 'Feminino':
        return 0
    else:
        return 2

data['sexo'] = data['sexo'].apply(mudarSexo)
data
    
```

```python
data.describe().transpose()
```

Criando as váriaveis e eliminando as colunas

```python
data['Burnout_cinismo'] = (data['ci1']+data['ci2']+data['ci3']+data['ci4'])/4
data['Burnout_exaustao'] = (data['ee1']+data['ee2']+data['ee3']+data['ee5']+data['ee6'])/6
data['Burnout_eficacia'] = (data['et1']+data['et2']+data['et3']+data['et4']+data['et5']+data['satXet6'])/6
data['instabilidade_equipe'] = (data['equipe1']+data['equipe2']+data['equipe3']+data['equipe4'])/4
data['instabilidade_tecnica'] = (data['tec1']+data['tec2']+data['tec3']+data['tec4'])/4
data['instabilidade_tarefa'] = (data['reque1']+data['reque2']+data['reque3']+data['req4']+data['req5']+data['req6']+data['req7'])/7


labels = ['S1', 'S2', 'S3',
          'C2', 'C3', 'C4', 'C1', 'EM4', 'EM1', 'EM3', 'EM2', 'T3', 'T1', 'T4',
          'T2', 'I2', 'I1', 'I4', 'I3', 'equipe1',
          'reque1', 'reque2', 'reque3', 'tec1', 'req4', 'tec2', 'tec3', 'tec4',
          'req5', 'equipe2', 'equipe3', 'req6', 'tec5', 'tec6', 'req7', 'equipe4',
          'ci2', 'et6', 'ci3', 'satXet6',
          'ee6', 'ee2', 'et5', 'ci1', 'et4', 'ee1', 'et1', 'et2', 'et3', 'ee3',
          'ci4', 'ee5', 'sat1', 'sat2']
data.drop(columns = labels, inplace=True)

```
